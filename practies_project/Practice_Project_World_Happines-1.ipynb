Data Description 

#The World Happiness Report is a landmark survey of state of global happiness. It describe how measurements of wel-being can be used effectively to assess the progress of nation. The happiness score and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation question asked in poll. This questions known as the Cantril ladder with best possible score is 10 and worst is 0. 

World Happines report consists of six columns like Economic production, Social support, Life expectancy, Freedom, Absence of corruption and Generosity. It also consist of hypothetical country named Dystopia that has values equal to the world's lowest national averages for each of the six factors.
